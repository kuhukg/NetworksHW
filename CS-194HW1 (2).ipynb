```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx

G = nx.karate_club_graph()
```

```python
nx.draw_networkx(G, with_labels=True, node_color=['g'] + ['#0F0F0F'] * 29 + ['b'] + ['#0F0F0F'] * 3)
```

```python
nx.draw_networkx(G, with_labels=True)
plt.show()
```

```python
nx.draw_random(G)
plt.show()
nx.draw_spectral(G)
plt.show()
nx.draw_circular(G)
plt.show()
nx.draw_spring(G)
plt.show()
```

```python
subNode = []
for i in range(0, 17): subNode.append(i)
sg = G.subgraph(subNode)
nx.draw_networkx(sg)
plt.show()
```

```python
import pickle 
fl = open("pickle.pkl", "wb")
pickle.dump(sg,fl)
fl.close()
fl = open("pickle.pkl", "rb")
sg2 = pickle.load(fl)
nx.draw_networkx(sg2)
plt.show()
```
