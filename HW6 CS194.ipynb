```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import numpy as np
import scipy as sp
from scipy import integrate
from pylab import *
import collections

```

```python
res_func = lambda x: (1-x)
f_func = lambda z: z
p_star = 0.15
y = [0]
z_i = 0
z = np.linspace(0,1, 100)
for i in range(1,100):
    y.append(res_func(p_star/f_func(z[i])))
scatter(z, y, marker='o')
show()
    
```

```python
res_func = lambda x: (1-x)
f_func = lambda z: z
p_star = 0.15
def plot_func(k):
    z = np.linspace(0,1, 100)
    y = [0]
    for i in range(1,100):
        n = z[i]
        for j in range (1,k):
            n = res_func(p_star/f_func(n))
        y.append(n)
    scatter(z, y, marker='o')
    show()
plot_func(2)
plot_func(10)
plot_func(50)
    
```

```python
z = np.linspace(0,1, 100)
rev = []
sur = []
res_func = lambda x: (1-x)*x
for i in z:
    new_func = lambda x: i
    if 4*i <= 1:
        curr = (-1 + np.sqrt(1-(4*i)))/(-2)
    rev.append(curr*i)
    demand = sp.integrate.quad(res_func, 0, curr)
    price = sp.integrate.quad(new_func, 0, curr)
    sur.append(demand[0] - price[0])

plt.plot(z, rev, label = "revenue")
plt.plot(z, sur, label = "surplus")
plt.legend()
    
    
```
